```python
def show(image):
    _, ax = pyplot.subplots(figsize=(14, 14))
    ax.axis("off")
    ax.imshow(image)
    pyplot.tight_layout()
```

```python
show(e.render())
```

```python
action = e.action_space.sample()
print(action)
```

```python
e.step(action)
```

```python
show(e.render())
```

```python
e._state
```

```python
def cross_target(left_right, high_start):
    move = left_right[0] if high_start else left_right[1]
    while high_start == (e._state[1] > e._state[0]):
        e.step(move)
        show(e.render())
```

```python
cross_target((0, 10), e._state[1] > e._state[0])
```

```python
cross_target((2, 8), e._state[1] > e._state[0])
```
