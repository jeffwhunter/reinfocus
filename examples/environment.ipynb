```python
def show(image):
    _, ax = pyplot.subplots(figsize=(14, 14))
    ax.axis("off")
    ax.imshow(image)
    pyplot.tight_layout()
```

```python
def cross_target(left, right):
    assert e._state is not None

    high_start = e._state[0, 1] > e._state[0, 0]

    move = left if high_start else right
    while high_start == (e._state[0, 1] > e._state[0, 0]):
        e.step(move)
        show(e.render())
```

```python
cross_target(0, 10)
```

```python
cross_target(2, 8)
```

```python
cross_target(4, 6)
```
