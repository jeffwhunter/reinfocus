```python
from matplotlib import pyplot

from reinfocus.environment import focus_environment
```

```python
e = focus_environment.SimpleDiscreteEnviroment(render_mode="rgb_array")
```

```python
e.action_space
```

```python
e.observation_space
```

```python
e.reset()
```

```python
def show(image):
    _, ax = pyplot.subplots(figsize=(14, 14))
    ax.axis("off")
    ax.imshow(image)
    pyplot.tight_layout()
```

```python
show(e.render())
```

```python
action = e.action_space.sample()
print(action)
```

```python
e.step(action)
```

```python
show(e.render())
```

```python
e._state
```

```python
def cross_target(right_left, high_start):
    move = right_left[1] if high_start else right_left[0]
    while high_start == (e._state[1] > e._state[0]):
        e.step(move)
        show(e.render())
```

```python
cross_target((5, 6), e._state[1] > e._state[0])
```

```python
cross_target((3, 4), e._state[1] > e._state[0])
```

```python
cross_target((1, 2), e._state[1] > e._state[0])
```
