```python
import matplotlib.pyplot as plt
import numpy as np
import reinfocus.graphics.render as ren
import reinfocus.graphics.world as wor
import reinfocus.vision as vis
```

```python
focus_values = []
focus_space = np.linspace(.1, 10, num=100)

world = wor.one_rect_world(distance=5)

for focus_distance in focus_space:
    print(f'Measuring {focus_distance}', end='\r')

    output = ren.render(
        focus_distance=focus_distance,
        world=world)

    focus_values.append(vis.focus_value(output))

plt.plot(focus_space, focus_values)
plt.show()
```
